# einfache Basismethode um python mit Datenbank zu verbinden, für Dataframe besser siehe weiter unten

```python
# https://www.freecodecamp.org/japanese/news/connect-python-with-sql/
def create_server_connection(host_name, user_name, user_password):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
```

```python


connection = create_server_connection("localhost", "projekt", "x")

cursor=connection.cursor()
cursor.execute("select distinct(city) from test.`sample-superstore`" )
r = cursor.fetchall()

r
```
